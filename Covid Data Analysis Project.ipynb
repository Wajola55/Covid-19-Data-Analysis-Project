# 3 Reshaping Data

### 3.1  Deleting columns and displaying each country name only once along with the earliest and latest dates for that country:

```python
# deleting columns from Covid-19 Global Dataset:

df_covid.drop(["WHO_region", "Country_code"], inplace = True, axis=1)


# Group the data by the 'Country' column and aggregate the 'Date_reported' column
date_range = df_covid.groupby('Country')['Date_reported'].agg(['min', 'max'])



# Merge the date range dataframe with the original dataframe on the 'Country' column
merged_df = pd.merge(df_covid, date_range, on='Country')


# Delete the 'Date_reported' column
merged_df.drop(columns=['Date_reported'], inplace=True)


# Move the 'min' and 'max' columns to the beginning of the dataframe
cols = merged_df.columns.tolist()
cols = cols[-2:] + cols[:-2]
merged_df = merged_df[cols]


# Drop duplicates from the 'Country' column
merged_df.drop_duplicates(subset=['Country'], inplace=True)



# Group the data by the 'Country' column and aggregate the columns
grouped_df = df_covid.groupby('Country').agg({
    'Date_reported': ['min', 'max'],
    'Cumulative_cases': 'sum',
    'New_deaths': 'sum',
    'Cumulative_deaths': 'sum'
})

# Reset the column names
grouped_df.columns = ['min', 'max', 'Cumulative_cases', 'New_deaths', 'Cumulative_deaths']

# Reset the index
grouped_df.reset_index(inplace=True)


grouped_df
```

### 3.2 Cleaning and aggregating the Covid-19 World Vaccination Progress dataset by removing unnecessary columns, grouping the data by country and aggregating the date range, and displaying each country name only once with the earliest and latest dates:

```python
# deleting columns from Covid-19 World Vaccination Progress:


df_vaccination.drop(["iso_code", "daily_vaccinations_raw", "daily_vaccinations", "total_vaccinations_per_hundred", 
                     "people_vaccinated_per_hundred", "people_fully_vaccinated_per_hundred", 
                     "daily_vaccinations_per_million", "source_name", "source_website"], inplace=True, axis=1)


# Group the data by the 'country' column and aggregate the 'date' column
date_range_vac = df_vaccination.groupby('country')['date'].agg(['min', 'max'])



# Merge the date range dataframe with the original dataframe on the 'Country' column
merged_df_vac = pd.merge(df_vaccination, date_range_vac, on='country')


# Delete the 'date' column
merged_df_vac.drop(columns=['date'], inplace=True)


# Move the 'min' and 'max' columns to the beginning of the dataframe
cols = merged_df_vac.columns.tolist()
cols = cols[-2:] + cols[:-2]
merged_df_vac = merged_df_vac[cols]


# Drop duplicates from the 'Country' column
merged_df_vac.drop_duplicates(subset=['country'], inplace=True)


merged_df_vac
```

#                                                4 Data Visualization

### 4.1 COVID-19 Cumulative Deaths by Country(Top 20):

```python

# Sort the countries by cumulative deaths and select the top 20
top_countries = grouped_df.sort_values(by='Cumulative_deaths', ascending=False).head(20)

# Create figure and axis objects
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(1, 1, 1)

# Set x-axis scale to logarithmic
ax.set_xscale('log')

# Create bar plot
x = sns.barplot(x='Country', y='Cumulative_deaths', data=top_countries, linewidth=2.5, ax=ax)

# Set x-axis tick labels to country names
ax.set_xticklabels(top_countries['Country'], rotation=90)

# Add x-axis label
plt.xlabel('Country')

# Add y-axis label
plt.ylabel('Cumulative Deaths')

# Add plot title
plt.title('COVID-19 Cumulative Deaths by Country (Top 20)')

# Display plot
plt.show()

```

### 4.2 COVID-19 People Fully Vaccinated by Country (Top 20):

```python
# Sort the countries by people fully vaccinated and select the top 20
top_vaccinated = merged_df_vac.sort_values(by='people_fully_vaccinated', ascending=False).head(20)

# Create figure and axis objects
fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(1, 1, 1)

# Create bar plot
x = sns.barplot(x='country', y='people_fully_vaccinated', data=top_vaccinated, ax=ax)

# Set x-axis tick labels to country names
ax.set_xticklabels(top_vaccinated['country'], rotation=90)

# Add x-axis label
plt.xlabel('Country')

# Add y-axis label
plt.ylabel('People Fully Vaccinated')

# Add plot title
plt.title('COVID-19 People Fully Vaccinated by Country (Top 20)')

# Display plot
plt.show()
```
